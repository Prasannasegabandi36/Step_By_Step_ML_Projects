```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url='https://raw.githubusercontent.com/das-amlan/Forest-Fire-Prediction/refs/heads/main/Algerian_forest_fires_dataset.csv'
df=pd.read_csv(url)
df.head(4)
```

```python
df.info()
```

#Data Cleaning


```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1

#df=dataset
```

```python
df.head()
```

```python
df[['Region']]=df[["Region"]].astype(int)
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
#fix space in col
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

change other columns into float data datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='0']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

```python
df.head()
```

```python
#lets have clean dataset
df.to_csv(url,index=False)
```
